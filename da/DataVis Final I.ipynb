```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

vi='Very interested'
si='Somewhat interested'
ni='Not interested'

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
total_respondents = 2233

bar_colors={vi: '#5cb85c', si: '#5bc0de', ni: '#d9534f' }
```

```python
df = pd.read_csv('./data/Topic_Survey_Assignment.csv')
```

```python
df
```

```python
df=round(100 * df / total_respondents, 2)
plt.figure(figsize=[20,8])

labels = list(df.index)

x = np.arange(len(labels))  # the label locations
x *= 5

width = 0.8  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df[vi], width, label=vi, color=bar_colors[vi])
rects2 = ax.bar(x + width/2, df[si], width, label=si, color=bar_colors[si])
rects3 = ax.bar(x + width * 1.5, df[ni], width, label=ni, color=bar_colors[ni])

ax.legend()
ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', {'fontsize': 16})
ax.set_xticks(x + width/2)
ax.set_xticklabels(labels, {'fontsize': 14}, rotation='vertical')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()
plt.subplots_adjust(bottom=0.3)
```

```python
plt.show()
```
